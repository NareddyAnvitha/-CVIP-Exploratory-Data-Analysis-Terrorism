```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/terrorism1.csv',encoding='latin1')
print(df)
```

```python
df.head()
```

```python
df['iyear'].value_counts()
```

```python
plt.figure(figsize=(10,3))
sns.barplot(x='iyear',y='iyear',data=df)
plt.xticks(rotation=25)
plt.xlabel('Attack year')
plt.ylabel('Number of attack each year')
plt.title('Attack of years')
```

```python
df['country_txt'].value_counts
```

```python
df['imonth'].value_counts()
plt.figure(figsize=(10,3))
df['imonth'].value_counts().sort_index().plot(kind='bar')
plt.xticks(rotation=25)
plt.xlabel('favoriate attacking month')
plt.ylabel('Mumber of attack on each month')
plt.title('most attack in the month')
```

```python
df['iday'].value_counts()
plt.figure(figsize=(10,3))
df['iday'].value_counts().sort_index().plot(kind='bar')
plt.xticks(rotation=25)
plt.xlabel('favorite attacking day')
plt.ylabel('Number of attack on each day')
plt.title('most attack in the day')
```

```python
plt.figure(figsize=(10,3))
sns.barplot(x=df['country_txt'].value_counts().head(25).index,palette='winter_r',y=df['country_txt'].value_counts().head(25).values)
plt.xticks(rotation=25)
plt.xlabel("attack on country")
plt.ylabel('Number of attack on each country')
plt.title('most attack in the country')
```

```python
df['region'].value_counts().max
plt.figure(figsize=(10,3))
df['region'].value_counts().plot.pie(shadow=True,autopct='%1.2f%%')
plt.xticks(rotation=25)
plt.xlabel('attack on particular region')
plt.ylabel('Number of attack on each region')
plt.title('most attack in the region')
```
